# Formato Arquivo DNA





O arquivo de origem deve estar no formato .csv, se chamar "origem.csv" e possuir os dados dispostos da seguinte forma: 

1ª Coluna:
RSID: string
Se disponível, carrega a variante rsID, que identifica o ramo dentro do banco de dados genético público dbSNP(SNP, sigla em inglês para Polimorfismo de Nucleotídeo Único, é uma variação do DNA em uma única base nucleotídica que nos permite traçar propensões e características..)

2ª Coluna:
CHROMOSOME: string
Indica o número do cromossomo onde se situa a variante genética. Existem 22 pares de cromossomos autossômicos que são indicados com um número e os cromossomos sexuais que são nomeados como X e Y. Há também marcadores no DNA mitocondrial que aparecem listados como “MT”.

3ª Coluna:
POSITION: string
É a posição variante no cromossomo, a posição do marcador genético de acordo com o  conjunto genético humano GRCh37 (http://grch37.ensembl.org/index.html).

4ª Coluna:
RESULT: string
• O par de valores observados (alelo 1 e alelo 2 - representando os lados paterno e materno, não necessariamente nessa ordem), contém resultados de genotipagem para os marcadores genéticos.

# Carregando Bancos de Dados e o Comparador:

```python
doencas_df = pd.read_csv('https://raw.githubusercontent.com/MuriloKrominski/DNA-X-Doencas-e-Predisposicoes-Geneticas/main/doencas.csv', sep=",", encoding='utf-8', low_memory="False")
doencas_df['RSID+CHROMOSOME+RESULT']= doencas_df['RSID+CHROMOSOME+RESULT'].values.astype(str)

origem_df = pd.read_csv('/content/origem.csv', sep=",", encoding='utf-8', low_memory="False")

origem_df = origem_df.drop(columns=['POSITION'])

origem_df['RSID']= origem_df['RSID'].values.astype(str)
origem_df['CHROMOSOME']= origem_df['CHROMOSOME'].values.astype(str)
origem_df['RESULT']= origem_df['RESULT'].values.astype(str)

origem_df['RSID+CHROMOSOME+RESULT'] = origem_df[['RSID','CHROMOSOME','RESULT']].apply(' '.join, axis=1)

origem_df = origem_df.drop(['RSID','CHROMOSOME','RESULT'],axis=1)

resultado_df = pd.merge(origem_df, doencas_df, 
                   on=['RSID+CHROMOSOME+RESULT'], 
                   how='inner')

resultado_df = resultado_df.sort_values(['DISEASE'])

resultado_df = resultado_df.reset_index(drop=True)
```
